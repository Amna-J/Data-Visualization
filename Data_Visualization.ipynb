```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
column_name='species'

plt.figure(figsize=(8,6))
plt.hist(data[column_name],bins=30,color='blue',edgecolor='black')
plt.title(f'IRIs Flower Histogram of {column_name}')
plt.xlabel(column_name)
plt.ylabel('frequency')
plt.grid(False)
plt.show()
```

```python
#plot histograms without grid
axes=data.hist(bins=30, figsize=(15,10),color='green', edgecolor='black')
plt.suptitle('Histogram of the entire IRIS data',fontsize=16)
plt.tight_layout()

#Remove grids from each subplot
for ax in axes.flatten():
  ax.grid(False)

plt.show()
```

```python
x_col="sepal_length"
y_col="petal_length"

plt.figure(figsize=(8,6))  #control graph size
plt.scatter(data[x_col], data[y_col], c="orange",s=40,alpha=0.7,edgecolors="black")

plt.title("Scatter Plot", fontsize=14)
plt.xlabel(x_col,fontsize=12)
plt.ylabel(y_col,fontsize=12)
plt.show()
```

```python
x_col="species"   #categorical column
y_col="sepal_length"    #numerical column

plt.figure(figsize=(8,6))  #control graph size
sns.boxplot(x=x_col, y=y_col, hue=x_col,data=data,palette="pastel",legend=True)

plt.title("Box Plot", fontsize=14)
plt.xlabel(x_col,fontsize=12)
plt.ylabel(y_col,fontsize=12)
plt.show()
```

```python
x_col="species"   #categorical column
y_col="sepal_length"    #numerical column

plt.figure(figsize=(8,6))  #control graph size
sns.barplot(x=x_col, y=y_col, hue=x_col,data=data,palette="pastel",legend=False)

plt.title("Bar Plot", fontsize=14)
plt.xlabel(x_col,fontsize=12)
plt.ylabel(y_col,fontsize=12)
plt.show()
```
